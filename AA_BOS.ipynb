```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Perform seasonal decomposition
result = seasonal_decompose(data['Passengers_Domestic'], model='additive', period=12)
result.plot()
plt.show()

# Prepare data for SARIMA model
train = data['Passengers_Domestic'][:len(data)-12]
test = data['Passengers_Domestic'][-12:]

# Fit SARIMA model
model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit()

# Make predictions
forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Test Data')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, label='Forecast')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='k', alpha=0.2)
plt.title('SARIMA Forecast for Domestic Passengers')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test, forecast.predicted_mean))
print(f'RMSE: {rmse}')
```

### Feature Engineering

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, shuffle=False)
```

### Passanger Domestic

```python
from pmdarima import auto_arima

# ARIMA model
model = auto_arima(train_data['Passengers_Domestic'], seasonal=True, m=12)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Make predictions on the test data
predictions = model.predict(n_periods=len(test_data))

# Calculate evaluation metrics
mae = mean_absolute_error(test_data['Passengers_Domestic'], predictions)
rmse = mean_squared_error(test_data['Passengers_Domestic'], predictions, squared=False)

print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')

# Visualize the predictions
plt.figure(figsize=(12, 6))
plt.plot(test_data['Date'], test_data['Passengers_Domestic'], label='Actual')
plt.plot(test_data['Date'], predictions, label='Predicted')
plt.title('Domestic Passenger Count Forecast')
plt.xlabel('Date')
plt.ylabel('Passenger Count')
plt.legend()
plt.show()
```

### Passanger International

```python
from pmdarima import auto_arima

# Fit an ARIMA model
model = auto_arima(train_data['Passengers_International'], seasonal=True, m=12)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Make predictions on the test data
predictions = model.predict(n_periods=len(test_data))

# Calculate evaluation metrics
mae = mean_absolute_error(test_data['Passengers_International'], predictions)
rmse = mean_squared_error(test_data['Passengers_International'], predictions, squared=False)

print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')

# Visualize the predictions
plt.figure(figsize=(12, 6))
plt.plot(test_data['Date'], test_data['Passengers_International'], label='Actual')
plt.plot(test_data['Date'], predictions, label='Predicted')
plt.title('International Passenger Count Forecast')
plt.xlabel('Date')
plt.ylabel('Passenger Count')
plt.legend()
plt.show()
```
